```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,10,0.5)
y = np.sin(x)
plt.plot(x,y)
plt.title("Sinüs Grafiği")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(70)
y = np.random.rand(70)
colors = np.random.rand(70)
sizes = np.random.randint(61,170,70)

plt.scatter(x,y,c = colors,s = sizes)
plt.title("Nokta Grafisi")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(300)
plt.hist(data,bins = 125)

plt.title("Histogram")
plt.xlabel("Degereler")
plt.ylabel("Frequency")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ['A','B','C','D','E']
y = np.random.randint(1,10,5)
plt.bar(x,y)

plt.title("Sinüs Grafisi")
plt.xlabel("Kategoriler")
plt.ylabel("Değerler")

plt.show()
```
